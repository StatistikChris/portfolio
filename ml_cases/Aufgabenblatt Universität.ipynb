a) Write a function t_test(X, μ1, α) which tests the hypothesis
H0 :μ=μ1 againstH1 :μ≠μ1
for some given level α and a given vector of observations X . Your function should return 1 if you can reject the null hypothesis and 0 otherwise.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import math
```

```python
def t_test(x, mu0, alpha):
    
    decision = 0
    quantile = stats.t.ppf(q = 1-alpha/2 , df=len(x)-1)
    T = abs(ma.sqrt(len(x))*(np.mean(x)-mu0)/np.std(x, ddof=1)) # Teststatistik
    if T >= quantile: # Nullhypothese verwerfen, falls Teststatistik größer als das Quantil
        decision = 1
    
    return decision
    
```

e) Write a function t_test_onesided(X, μ1, α, alternative) to test the hypotheses from 4.d).


```python
def t_test_onesided(x, mu0, alpha, alternative):
    
    decision = 0
    left = 0
    
    if alternative == 'left': # hier schaut die Funktion obs linksseitig oder rechtsseitig ist und ....
        left = 1
        
    quantile = stats.t.ppf(1-alpha , len(x)-1)
        
    T = ma.sqrt(len(x))*(np.mean(x)-mu0)/np.std(x, ddof=1)
    
    if (-1)**left*T >= quantile: # ... jetzt muss halt eben entweder mit dem positiven Wert oder dem negativen Wert 
        # vergleichen werden
        decision = 1
    
    return decision
```

```python
def partialling_out(X,y,part_var):
    W = X.drop(np.array([part_var]), axis=1)
    modelpout1 = sklearn.linear_model.LinearRegression()
    modelpout1.fit(W, y)
    yhat = modelpout1.predict(W)
    resid1 = y-yhat
    modelpout2 = sklearn.linear_model.LinearRegression()
    modelpout2.fit(W, X[[part_var]])
    Xhat = modelpout2.predict(W)
    resid2 = X[[part_var]]-Xhat
    modelpout3 = sklearn.linear_model.LinearRegression()
    modelpout3.fit(resid2, resid1)
    predpout =  modelpout3.predict(resid2)
    beta = modelpout3.coef_
    
    return [resid1,resid2,predpout,beta]
    
    
```

c) Set up a simulation study to illustrate that the partialling out estimator is asymptotically normally distributed around the true value β. Repeat the estimation R = 1000 times.

```python
# Um zu zeigen, dass die Beta´s der Regressionen asymptotisch Normalverteilt sind, um das wahre Beta, werden 10000 
# zufällige Regressionen gerechnet und sich die Beta´s dann im Histogram angeschaut.
R = 10000
estimates = []
```

```python
for i in range(R):
    
    X = sample_generator_fn(100,5)
    residuals = np.random.normal(0,1,100)
    y = 5*X['x1'] + 4*X['x2'] + 3*X['x3'] + 2*X['x4'] + 1*X['x5'] + 0 + residuals
    
    output = partialling_out(X,y,'x1')
    output = output[3]
    estimates = np.append(estimates,output)
    
```

where Ci,j contains the number of observations of the test sample, which are in class i and classified as
class j. Write a function which takes two vectors (where each entry is either zero or one) as input and calculates the confusion matrix. Use your function to calculate the confusion matrix on the test set for both classifiers.

```python
# das geht 1 zu 1 nach den angegebenen Formeln

def confusion_matrix_fn(truth,result):
    c11 = sum(truth*result)
    c10 = sum(truth*(1-result))
    c01 = sum((1-truth)*result)
    c00 = sum((1-truth)*(1-result))
    tab = np.matrix([[c11, c10], [c01, c00]])
    return tab
```

g) Finally we combine precision and recall to the F1 -score
F1:= 2 . 1+1
precision recall
Write a function which takes two vectors as input and calculates the with different tuning parameters and maximize the F1 -score.

```python
# das geht 1 zu 1 nach den angegebenen Formeln

def F_score_fn(truth,result):
    c11 = sum(truth*result)
    c10 = sum(truth*(1-result))
    c01 = sum((1-truth)*result)
    prec = c11/(c11+c01)
    rec = c11/(c11+c10)
    F=2/(1/prec+1/rec)
    
    return F
```
